```python
%reset -f
import numpy as np
from scipy.io import loadmat, savemat
import matplotlib.pyplot as plt
import os
```

```python
BGN = 2
Z = 52
checkStr = '_BG_' + str(BGN) + '_Z_' + str(Z) + '.mat'
```

```python
folder_path = 'result/'
items = os.listdir(folder_path)
```

```python
PER = ()
SNR = ()
for i in range(0,len(items)):
    filename = folder_path + items[i]
    if filename[-14:]==checkStr :
        MatData = loadmat(filename)
        SNR_file = float(filename[11:-14])
        Pe_file = MatData['PeMC'][0,0]
        PER = PER + (Pe_file,)
        SNR = SNR + (SNR_file,)
    
```

```python
SortIdx = np.argsort(SNR)
SNR = np.array(SNR)[SortIdx]
PER = np.array(PER)[SortIdx]
```

```python
plt.semilogy(SNR,PER,'-o')
plt.xlabel('SNR (dB)')
plt.ylabel('PER')
plt.grid(True)
plt.title('Monte Carlo Simulation Using Python')
# plt.show()
plt.savefig('result\PER_SNR_Python.pdf')
```
